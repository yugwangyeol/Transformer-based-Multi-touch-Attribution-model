## Case report_1 

```python
file_id = 1
```

```python
adpk_bid_success_1 = pd.read_csv(path+"/adpk/bid_success/"+adpk_bid_success_files[file_id])[['date_time','log_type','adid']]
adpk_click_files_1 = pd.read_csv(path+"/adpk/click/"+adpk_click_files[file_id])[['date_time','log_type','adid']]
adpk_conversion_1 = pd.read_csv(path+"/adpk/conversion/"+adpk_conversion_files[file_id])[['date_time','log_type','adid']]
```

```python
commerce_click_1 = pd.read_csv(path+"/commerce/click/"+commerce_click_files[file_id])[['date_time','event_nm','adid']]
commerce_purchase_1 = pd.read_csv(path+"/commerce/purchase/"+commerce_purchase_files[file_id])[['date_time','event_nm','adid']]
commerce_search_1 = pd.read_csv(path+"/commerce/search/"+commerce_search_files[file_id])[['date_time','event_nm','adid']]
commerce_shoppingbag_1 = pd.read_csv(path+"/commerce/shoppingbag/"+commerce_shoppingbag_files[file_id])[['date_time','event_nm','adid']]

commerce_click_1.columns = ['date_time','log_type','adid']
commerce_purchase_1.columns = ['date_time','log_type','adid']
commerce_search_1.columns = ['date_time','log_type','adid']
commerce_shoppingbag_1.columns = ['date_time','log_type','adid']
```

```python
df_1 = pd.concat([adpk_bid_success_1,adpk_click_files_1,adpk_conversion_1,commerce_click_1,commerce_purchase_1,commerce_search_1,commerce_shoppingbag_1])
```

```python
df_1 = df_1.sort_values('date_time')
df_1
```
